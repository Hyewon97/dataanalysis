### at
이름으로 한 위치의 값만 가져온

```python
# data.at[0,'tip']    # ok
# data.loc[0,'tip']   # ok
# data.iloc[0,1]      # ok
# data.at[0]['tip']   #TypeError
# data.loc[0]['tip']  # ok
# data.iloc[0]['tip'] # ok
data['tip'] # 열 지정. 하나만 적어준다.
data[3:5]    # 행 지정. n번에 있는 것을 가지고 와라,, 범위 지정 가능(슬라이싱 이용)
```

```python
data
```

```python
# at은 slicing을 사용할 수 없다.
# data.at[0:3,'tip'] # 슬라이싱을 사용할 수 없다.
```

```python
# data['tip','sex'] # 인덱싱 방법으로 못가지고 온다.
# data[['tip','sex']] # 이렇게는 가지고 올 수 있다.

# iloc는 슬라이싱, 해싱 인덱싱을 사용할 수 있다.

data.iloc[:,1] # 인덱스를 이용해 가지고 온다, 1열에 있는 값을 가지고 온다.
data.iloc[:,[1,2]] # 펜싱 기법이여서 [[안에 :을 사용하지 못한다.]], 떨어져 있는 경우 이 방법을 사용한다.
data.iloc[:,1:3] 
data.loc[:,'tip':'sex'] # 이름을 잉요해서 가지고 올 수 있다. 떨어져 있으면 이 방법을 사용하지 못한다.
data.iloc[0,1] # 특정 위치에 있는 값을 가지고 온다.

# iloc loc 차이점은 이름 또는 인덱스를 가지고 와야하느것, 지정할때 행이 먼저 오고 그 담에 열이 오고.
# at은 특정 위치에 있는 것을 가지고 와야 하기 때문에 행렬을 가지고 오지만, 이름으로 가지고 와서 하나씩만 가지고 온다?
# 데이터 프레임을 가지고 올 때, 행 또는 열, 하나씩만 가지고 올 수 있다.
# data[열] # 열 하나만 지정
# data[행:행] # 행의 범위를 지정
# data[['tip','day']] #ok
# data['tip':'day']   #TypeError   # 슬라이싱 사용시 오류 발생. 슬라이싱은 행일때만 사용한다.
```
