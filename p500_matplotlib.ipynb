```python
# 선 그래프
plt.plot(x,y)
```

```python
# 선 그래프
#(x값, y값, linewidth 선의 두께, c 선의 색상, alpha 투명도),,, 키워드를 사용하기 때문에 순서는 상관없다.
plt.plot(x,y, linewidth=5, c='r', alpha=0.5, label='line graph')


# scatter(점) 그래프
plt.scatter(x,y,s=150, c='g',alpha=0.5, label='dot graph')

# 격자모양
plt.grid()


# x축, y축의 범위 지정
plt.xlim([0,6])
plt.ylim([0,12])


# 범례
# plt.scatter(),plt.plot() 함수에서 label 속성을 지정하지 않으면 아래와 같은 메세지가 출력된다.
# No handles with.. 어쩌구 노침
# plt.legend()


plt.legend(loc=4, fontsize=15) # 값을 지정하지 않으면 범례가 알아서 적절한 곳에서 출력된다.
# plt.legend(loc='lower right', fontsize=15)

# ticks는 각 눈금에 표시된 숫자들로 이들의 크기 및 색상을 설정
plt.xticks(fontsize=14, color='b')
plt.yticks(fontsize=16, color='y')

# 제목
plt.title('y=2x', fontsize=20, c='m')

# 축에 대한 설명
plt.xlabel('Sun light', fontsize='15', c='b')
plt.ylabel('Apple Sweat', fontsize='15', color='y')

# 수직, 수평 그릭
plt.axvline(x=2, linewidth=10) # 수직
plt.axhline(y=4,linewidth=8) # 수평

plt.show()
```
